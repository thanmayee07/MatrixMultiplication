###GEMM Sequential vs Multi-threaded implementation###

```python
import numpy as np
import matplotlib.pyplot as plt

# Define matrix sizes and density
matrix_sizes = [2**i for i in range(7, 12)]  # Matrix sizes ranging from 218 to 2048
density = 0.1  # Constant density

# Manually input execution times for sequential and multi-threaded CPU
# Each element corresponds to a matrix size
sequential_cpu_times = [8.717000, 16.571000, 56.607000, 418.136000, 5770.316000]  # Execution times for sequential CPU
multi_threaded_cpu_times = [0.004710, 0.031675, 0.310484, 1.865056, 34.789377]  # Execution times for multi-threaded CPU

# Plotting CPU execution times
plt.plot(matrix_sizes, sequential_cpu_times, label='Sequential CPU', marker='o')
plt.plot(matrix_sizes, multi_threaded_cpu_times, label='Multi-threaded CPU', marker='x')

plt.title('Performance Comparison of Sequential CPU and Multi-threaded CPU')
plt.xlabel('Matrix Size')
plt.ylabel('Execution Time (ms)')
plt.legend()
plt.grid(True)
plt.xscale('log', base=2)  # Set logarithmic scale for x-axis
plt.xticks(matrix_sizes, [str(size) for size in matrix_sizes])  # Set x-axis tick labels
plt.show()
```

###GEMM Sequential vs GPU###


```python
import numpy as np
import matplotlib.pyplot as plt

# Define matrix sizes and density
matrix_sizes = [2**i for i in range(7, 12)]  # Matrix sizes ranging from to 128 to 2048
density = 0.1  # Constant density

# Manually input execution times for sequential and multi-threaded CPU
# Each element corresponds to a matrix size
sequential_cpu_times = [8.717000, 16.571000, 56.607000, 418.136000, 5770.316000]  # Execution times for sequential CPU
gpu_times = [0.250656, 0.351712, 0.842368, 5.388512, 41.635777]  # Execution times for multi-threaded CPU

# Plotting CPU execution times
plt.plot(matrix_sizes, sequential_cpu_times, label='Sequential CPU', marker='o')
plt.plot(matrix_sizes, gpu_times, label='GPU', marker='x')

plt.title('Performance Comparison of Sequential CPU and GPU')
plt.xlabel('Matrix Size')
plt.ylabel('Execution Time (ms)')
plt.legend()
plt.grid(True)
plt.xscale('log', base=2)  # Set logarithmic scale for x-axis
plt.xticks(matrix_sizes, [str(size) for size in matrix_sizes])  # Set x-axis tick labels
plt.show()
```

### SpMV performance metrics on GPU ###


```python
import numpy as np
import matplotlib.pyplot as plt

# Define densities and constant matrix size
densities = [0.1, 0.3, 0.5, 0.7, 1]  # Varying densities
matrix_size = 2**15  # Constant matrix size

# Manually input execution times, GFLOPS, and throughput for GPU
# Each element corresponds to a density
gpu_times = [10.773568, 10.816576, 9.581696, 8.336576, 7.558496]  # Execution times (ms)
gpu_gflops = [19.932892, 59.560911, 99.791917, 218.368754, 284.115206]  # GFLOPS
gpu_throughput = [19932891772.035614, 59560910935.222900, 99791916848.981018, 218368754208.505615, 284115206048.867004]  # Throughput (OPS)

# Plotting GPU execution times, GFLOPS, and throughput for different densities
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.plot(densities, gpu_times, marker='o', label='Execution Time')
plt.xlabel('Density')
plt.ylabel('Execution Time (ms)')
plt.title('GPU Execution Time')
plt.grid(True)

plt.subplot(1, 3, 2)
plt.plot(densities, gpu_gflops, marker='x', label='GFLOPS')
plt.xlabel('Density')
plt.ylabel('GFLOPS')
plt.title('GPU GFLOPS')
plt.grid(True)

plt.subplot(1, 3, 3)
plt.plot(densities, gpu_throughput, marker='s', label='Throughput')
plt.xlabel('Density')
plt.ylabel('Throughput (OPS)')
plt.title('GPU Throughput')
plt.grid(True)

plt.tight_layout()
plt.show()
```
